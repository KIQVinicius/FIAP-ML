```python
import pandas as pd
mercado = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/data/main/RiscoCredito%20-%20hot_enc.csv')
mercado
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
association_rules(apriori(mercado,min_support=0.20,use_colnames=True))
```
